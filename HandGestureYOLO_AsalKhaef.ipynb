```python
import os
import random


random.seed(42)

# Define source folders
folders = ["/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S1","/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S", "/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S3", "/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S4", "/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S5"]

train_ratio = 0.8

train_files = []
val_files = []

for folder in folders:
    subjects = [sub for sub in os.listdir(folder) if os.path.isdir(os.path.join(folder, sub)) and sub.startswith("Subject")]
    for subject in subjects:
        image_dir = os.path.join(folder, subject)
        label_dir = os.path.join(folder, f"{subject}_yolo")

        images = sorted([os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith(".png")])
        labels = sorted([os.path.join(label_dir, f) for f in os.listdir(label_dir) if f.endswith(".txt")])



        # Shuffle and split
        data = list(zip(images, labels))
        random.shuffle(data)

        split_idx = int(len(data) * train_ratio)
        train_files += data[:split_idx]
        val_files += data[split_idx:]

# Display some samples
print(f"Train samples: {len(train_files)}, Validation samples: {len(val_files)}")

# Example: View a few training and validation samples
print("Sample Train Files:")
for img, lbl in train_files[:3]:
    print(f"Image: {img}, Label: {lbl}")

print("Sample Validation Files:")
for img, lbl in val_files[:3]:
    print(f"Image: {img}, Label: {lbl}")
```

```python
import os

def prepend_to_png_filenames(folder_path, string_to_prepend):
    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        # Check if it's a file and has a .png extension
        if os.path.isfile(file_path) and filename.endswith('.txt'):
            # Create the new file name by prepending the string
            new_filename = string_to_prepend + filename
            new_file_path = os.path.join(folder_path, new_filename)

            os.rename(file_path, new_file_path)

folder_path = '/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S1/Subject1_yolo'
string_to_prepend = '1_'
prepend_to_png_filenames(folder_path, string_to_prepend)
```

```python
import os

# Define classes and their indices
classes =  ['Punch_VFR','One_VFR','Two_VFR','Three_VFR','Four_VFR','Five_VFR','Six_VFR','Seven_VFR','Eight_VFR','Nine_VFR','Span_VFR','Horiz_HFR','Collab','XSign','TimeOut']  # List of class names

class_column_index = {
    'Punch_VFR': 0,
    'One_VFR': 2,
    'Two_VFR': 4,
    'Three_VFR': 6,
    'Four_VFR': 8,
    'Five_VFR': 10,
    'Six_VFR': 12,
    'Seven_VFR': 14,
    'Eight_VFR': 16,
    'Nine_VFR': 18,
    'Span_VFR': 20,
    'Horiz_HFR': 25,
    'Collab': 26,
    'XSign': 27,
    'TimeOut': 28
}

# Input dataset file (update the path to your file)
input_file = "/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/Subject5.txt"
output_labels_dir = "/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/S5/Subject5"
output_base_file = "/content/drive/MyDrive/Hand_Gesture_YOLO_AsalKhaef/processed_dataset5.txt"  # Processed dataset file

# Ensure the output labels directory exists
os.makedirs(output_labels_dir, exist_ok=True)

def convert_to_yolo_annotation(input_file):
    with open(input_file, 'r') as file, open(output_base_file, 'w') as base_file:
        headers = file.readline().strip()
        base_file.write(headers + '\n')

        for line in file:
            line_data = line.strip().split(",")
            color_image_path = line_data[0]
            depth_image_path = line_data[1]
            bbox_data = line_data[2:]

            # Extract only the base filename (e.g., '316_color.png' from the full path)
            image_name = os.path.basename(color_image_path.replace("\\", "/"))
            base_name = image_name.replace('.png', '.txt')

            # Generate YOLO annotations for desired classes
            yolo_annotations = []
            for class_id, class_name in enumerate(classes):
                bbox = bbox_data[class_column_index[class_name]]

                bbox_values = bbox.strip("[]").split()
                if len(bbox_values) == 4:
                    if bbox_values == ['0', '0', '0', '0']:
                        continue

                    # Otherwise, process the bounding box
                    x, y, w, h = map(float, bbox_values)
                    image_width = 960
                    image_height = 540
                    yolo_x = (x + w / 2) / image_width
                    yolo_y = (y + h / 2) / image_height
                    yolo_w = w / image_width
                    yolo_h = h / image_height
                    yolo_annotations.append(f"{class_id} {yolo_x:.6f} {yolo_y:.6f} {yolo_w:.6f} {yolo_h:.6f}")


            label_file_path = os.path.join(output_labels_dir, "5_"+base_name)

            with open(label_file_path, 'w') as label_file:
                label_file.write("\n".join(yolo_annotations))


            base_file.write(line + '\n')

# Process the dataset
convert_to_yolo_annotation(input_file)

print("Processing complete. YOLO annotation files saved in:", output_labels_dir)
print("Processed dataset saved in:", output_base_file)
```
