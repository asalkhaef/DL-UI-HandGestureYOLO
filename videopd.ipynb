```python
from ultralytics import YOLO
import os
import cv2
```

```python
model = YOLO("best_ep.pt")
```

```python
def convert_frames_to_video(output_video, frames_folder):
    frame_rate = 30
    frame_files = os.listdir(frames_folder)
    first_frame = cv2.imread(os.path.join(frames_folder, frame_files[0]))
    height, width, _ = first_frame.shape

    size = (width, height)

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use 'XVID' or 'MJPG' for other formats
    out = cv2.VideoWriter(output_video, fourcc, frame_rate, size)

    # Write frames to video
    for frame_file in frame_files:
        path = os.path.join(frames_folder, frame_file)
        frame = cv2.imread(path)
        frame = cv2.resize(frame, size)
        out.write(frame)
```

```python
def predict_image(img, i):
    results = model(img)

    for result in results:
        result.save(filename=f"frames/f_{i}.jpg")
```

```python
video_path = "asal-final.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    raise Exception("Error: Could not open video.")

frame_count = 0
while True:
    frame_count += 1
    ret, frame = cap.read()

    # Break the loop if no frames are left to read
    if not ret:
        print("Video has ended or failed to load frames.")
        break



    predict_image(frame, frame_count)

cap.release()
```
